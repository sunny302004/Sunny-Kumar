List down the compounded annual growth rate (CAGR) in 4-wheeler
units for the top 5 makers from 2022 to 2024.

```python


# Filter data for 4-wheelers and extract relevant columns
data_4w = data[data['vehicle_category_y'] == '4-Wheelers'][
    ['fiscal_year', 'maker', 'electric_vehicles_sold_y']
]

# Aggregate total sales for each maker by fiscal year
sales_summary = data_4w.groupby(['maker', 'fiscal_year']).sum().reset_index()

# Pivot to create a maker-wise year-wise sales table
sales_pivot = sales_summary.pivot(index='maker', columns='fiscal_year', values='electric_vehicles_sold_y')

# Define a function to calculate CAGR
def calculate_cagr(start, end, periods):
    if pd.notna(start) and pd.notna(end) and start > 0:
        return ((end / start) ** (1 / periods)) - 1
    return None

# Calculate CAGR for the period 2022 to 2024
sales_pivot['CAGR_2022_2024'] = sales_pivot.apply(
    lambda row: calculate_cagr(row.get(2022), row.get(2024), 2), axis=1
)

# Select top 5 makers based on their CAGR values
top_makers_cagr = sales_pivot['CAGR_2022_2024'].nlargest(5).reset_index()

# Output the top 5 makers with their CAGR
print(top_makers_cagr)
top_makers_cagr.to_csv('top_5_makers_cagr.csv')
```

What are the peak and low season months for EV sales based on the
data from 2022 to 2024?

```python
data['fiscal_year'] = pd.to_numeric(data['fiscal_year'], errors='coerce')

data_filtered = data[data['fiscal_year'].isin([2022, 2023, 2024])].copy()


# Convert 'date' column to datetime for easier processing
data_filtered.loc[:, 'date'] = pd.to_datetime(data_filtered['date'], format='%d-%b-%y')


# Filter data for 4-Wheelers only
four_wheelers_data = data_filtered[data_filtered['vehicle_category_y'] == '4-Wheelers'].copy()

# Group data by year and month to calculate total EV sales
four_wheelers_data.loc[:, 'year_month'] = four_wheelers_data['date'].dt.to_period('M')
monthly_sales = four_wheelers_data.groupby('year_month')['electric_vehicles_sold_y'].sum().reset_index()

# Rename columns for clarity
monthly_sales.rename(columns={'electric_vehicles_sold_y': 'total_ev_sales'}, inplace=True)

# Add a separate 'year' and 'month' column for analysis
monthly_sales.loc[:, 'year'] = monthly_sales['year_month'].dt.year
monthly_sales.loc[:, 'month'] = monthly_sales['year_month'].dt.month

# Find peak and low sales months for each year
peak_low_sales = monthly_sales.groupby('year').apply(
    lambda x: pd.DataFrame({
        'Peak Month': [x.loc[x['total_ev_sales'].idxmax(), 'month']],
        'Peak Sales': [x['total_ev_sales'].max()],
        'Low Month': [x.loc[x['total_ev_sales'].idxmin(), 'month']],
        'Low Sales': [x['total_ev_sales'].min()]
    }).reset_index()
)

print(peak_low_sales)
# peak_low_sales.to_csv('peak_low_sales.csv', index=False)
```

What is the projected number of EV sales (including 2-wheelers and 4-
wheelers) for the top 10 states by penetration rate in 2030, based on the
compounded annual growth rate (CAGR) from previous years?

```python
# Filter data for both 2-wheelers and 4-wheelers
ev_sales_data = data[(data['vehicle_category_y'].isin(['2-Wheelers', '4-Wheelers'])) | (data['vehicle_category_x'].isin(['2-Wheelers', '4-Wheelers']))].copy()

# Calculate EV penetration rate (EV sales / Total vehicle sales) for each state
# corrected the line below to sum the two columns first.
ev_sales_data['penetration_rate'] = (
    (ev_sales_data['electric_vehicles_sold_y'].fillna(0) + ev_sales_data['electric_vehicles_sold_x'].fillna(0)) / ev_sales_data['total_vehicles_sold']
)

# Aggregate data to find total EV sales and penetration rate by state
state_sales = ev_sales_data.groupby('state').agg({
    'electric_vehicles_sold_y': 'sum',
    'electric_vehicles_sold_x': 'sum',
    'total_vehicles_sold': 'sum',
    'penetration_rate': 'mean'
}).reset_index()

# Sort by penetration rate and get the top 10 states
top_states = state_sales.sort_values(by='penetration_rate', ascending=False).head(10)

# Add historical sales data for CAGR calculation
historical_sales = ev_sales_data.groupby(['state', 'date']).agg({
    'electric_vehicles_sold_y': 'sum',
        'electric_vehicles_sold_x':'sum'
}).reset_index()


# Filter for the top 10 states only
top_states_sales = historical_sales[historical_sales['state'].isin(top_states['state'])].copy()


# Calculate CAGR for each state
def calculate_cagr(start_value, end_value, years):
    if start_value > 0:
       return ((end_value / start_value) ** (1 / years)) - 1
    else:
        return 0

# Define start and end years for CAGR calculation
top_states_sales['date'] = pd.to_datetime(top_states_sales['date'])
start_year = top_states_sales['date'].dt.year.min()
end_year = top_states_sales['date'].dt.year.max()

# Add CAGR calculations to the top states data
cagr_results = []
for state in top_states['state']:
    state_data = top_states_sales[top_states_sales['state'] == state]
    if len(state_data) >= 2:  # Ensure enough data for CAGR
        start_sales = (state_data[state_data['date'].dt.year == start_year][['electric_vehicles_sold_y','electric_vehicles_sold_x']].sum()).sum()
        end_sales = (state_data[state_data['date'].dt.year == end_year][['electric_vehicles_sold_y','electric_vehicles_sold_x']].sum()).sum()
        years = end_year - start_year
        cagr = calculate_cagr(start_sales, end_sales, years)
        cagr_results.append({'state': state, 'start_sales': start_sales, 'end_sales': end_sales, 'CAGR': cagr})

cagr_df = pd.DataFrame(cagr_results)

# Project sales for 2030
projected_sales = cagr_df.copy()
projected_sales['projected_sales_2030'] = projected_sales['end_sales'] * (
    (1 + projected_sales['CAGR']) ** (2030 - end_year)
)

projected_sales.to_csv('projected_sales_2030.csv', index=False)

```

Estimate the revenue growth rate of 4-wheeler and 2-wheelers
EVs in India for 2022 vs 2024 and 2023 vs 2024, assuming an average
unit price.

```python
# Define average prices
price_2wheeler = 85000
price_4wheeler = 1500000

# Filter data for 2-wheelers and 4-wheelers
data_2wheeler = data[data['vehicle_category_y'] == '2-Wheelers'].copy()
data_4wheeler = data[data['vehicle_category_y'] == '4-Wheelers'].copy()
data_2wheeler = data[data['vehicle_category_x'] == '2-Wheelers'].copy()
data_4wheeler = data[data['vehicle_category_x'] == '4-Wheelers'].copy()


# Calculate revenue for a given year
def calculate_revenue(df, year, price_2w, price_4w):
    df_year = df[df['fiscal_year'] == year].copy()
    if df_year.empty:
        return 0
    # Correctly filter for 2-wheelers using logical OR operator
    total_2w_sales = df_year[
        (df_year['vehicle_category_y'] == '2-Wheelers') | (df_year['vehicle_category_x'] == '2-Wheelers')
    ]['electric_vehicles_sold_y'].fillna(0).sum() +  df_year[
        (df_year['vehicle_category_y'] == '2-Wheelers') | (df_year['vehicle_category_x'] == '2-Wheelers')
    ]['electric_vehicles_sold_x'].fillna(0).sum()
     # Correctly filter for 4-wheelers using logical OR operator
    total_4w_sales = df_year[
        (df_year['vehicle_category_y'] == '4-Wheelers') | (df_year['vehicle_category_x'] == '4-Wheelers')
    ]['electric_vehicles_sold_y'].fillna(0).sum() +  df_year[
        (df_year['vehicle_category_y'] == '4-Wheelers') | (df_year['vehicle_category_x'] == '4-Wheelers')
    ]['electric_vehicles_sold_x'].fillna(0).sum()
    revenue = (total_2w_sales * price_2w) + (total_4w_sales * price_4w)
    return revenue

# Calculate revenue for 2022, 2023, and 2024
revenue_2022 = calculate_revenue(data, 2022, price_2wheeler, price_4wheeler)
revenue_2023 = calculate_revenue(data, 2023, price_2wheeler, price_4wheeler)
revenue_2024 = calculate_revenue(data, 2024, price_2wheeler, price_4wheeler)


# Calculate growth rate
def calculate_growth_rate(revenue_start, revenue_end):
    if revenue_start == 0:
        return None
    growth_rate = ((revenue_end - revenue_start) / revenue_start) * 100
    return growth_rate

# Calculate revenue growth rate 2022 vs 2024
growth_rate_2022_2024 = calculate_growth_rate(revenue_2022, revenue_2024)

# Calculate revenue growth rate 2023 vs 2024
growth_rate_2023_2024 = calculate_growth_rate(revenue_2023, revenue_2024)


# Create a DataFrame to store the results
results_df = pd.DataFrame({
    'Metric': [
        "Total Revenue in 2022 :",
        "Total Revenue in 2023 :",
        "Total Revenue in 2024 :",
        "Revenue Growth Rate (2022 vs 2024) :",
        "Revenue Growth Rate (2023 vs 2024) :"
    ],
    'Value': [
        f"{revenue_2022:,.2f} INR",
        f"{revenue_2023:,.2f} INR",
        f"{revenue_2024:,.2f} INR",
        f"{growth_rate_2022_2024:.2f}%",
        f"{growth_rate_2023_2024:.2f}%"
    ]
})

# Save the DataFrame to a CSV file
results_df.to_csv('revenue_growth_results.csv', index=False)
results_df
```
